```python
import textwrap

import numpy as np
import seaborn as sns
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
matplotlib.use('TkAgg')

matplotlib.rcParams['font.family'] = 'Times New Roman'
%matplotlib inline
```

```python
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel(r'./data_processed/complementary_features/mer.xlsx')

my_pal = {"3mer": "#EE7600", "4mer": "#FFCC66", "5mer":"#1ABCC2", "6mer":"#6F94CD"}

f, ax = plt.subplots(figsize=(19, 6))

sns.violinplot(x=df["Prediction models"], y=df["x"], data=df, palette=my_pal, width=0.95, linewidth=1.6)

plt.xticks(fontsize=27)
plt.yticks(fontsize=27)

plt.ylabel("AUC", fontsize=27)
plt.xlabel("", fontsize=27)
plt.show()
```
