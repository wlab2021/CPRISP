```python
import numpy as np
import pandas as pd
from sklearn.metrics import roc_curve

import matplotlib
import matplotlib.pyplot as plt
matplotlib.use('TkAgg')
from sklearn.metrics import roc_auc_score
#tensorflow
# f, ax = plt.subplots(figsize=(11, 9))

matplotlib.rcParams['font.family'] = 'Times New Roman'
%matplotlib inline
```

```python
preds_np = np.load(r"./data_processed/across_heterogeneous_cell/s-FASTKD2-K-H.npy", allow_pickle=True)
preds = pd.DataFrame(preds_np)

labels_np = np.load(r"./data_processed/across_heterogeneous_cell/l-FASTKD2-K-H.npy", allow_pickle=True)
labels = pd.DataFrame(labels_np)
```

```python
plt.figure(figsize=(10, 10))

auc_cda = roc_auc_score(labels.values[0], preds.values[0])
fpr_cda, tpr_cda, thresholds = roc_curve(labels.values[0], preds.values[0])
plt.plot(fpr_cda,tpr_cda,color='#FF2400',label='CPRISP(AUC= %0.4f)' % auc_cda)

"""CircSSNN"""

auc_ncp = roc_auc_score(labels.values[1], preds.values[1])
fpr_ncp, tpr_ncp, thresholds = roc_curve(labels.values[1], preds.values[1])
plt.plot(fpr_ncp,tpr_ncp,color='#8FBC8F',label='CircSSNN(AUC= %0.4f)' % auc_ncp)

"""HCRNet"""

auc_sim = roc_auc_score(labels.values[2], preds.values[2])
fpr_sim, tpr_sim, thresholds = roc_curve(labels.values[2], preds.values[2])
plt.plot(fpr_sim,tpr_sim,color='#6F94CD',label='HCRNet(AUC= %0.4f)' % auc_sim)

"""CRBPDL"""

auc_vgae = roc_auc_score(labels.values[3], preds.values[3])
fpr_vgae, tpr_vgae, thresholds = roc_curve(labels.values[3], preds.values[3])
plt.plot(fpr_vgae,tpr_vgae,color='#CA6FAC',label='CRBPDL(AUC= %0.4f)' % auc_vgae)

"""iCircRBP-DHN"""

auc_gcn = roc_auc_score(labels.values[4], preds.values[4])
fpr_gcn, tpr_gcn, thresholds = roc_curve(labels.values[4], preds.values[4])
plt.plot(fpr_gcn,tpr_gcn,color='#1ABCC2',label='iCircRBP-DHN(AUC= %0.4f)' % auc_gcn)


"""CRIP"""

auc_crp = roc_auc_score(labels.values[5], preds.values[5])
fpr_crp, tpr_crp, thresholds = roc_curve(labels.values[5], preds.values[5])
# plt.plot(fpr_crp,tpr_crp,color='#2AB34A',label='CRPGCN(Circ2Disease based on MNDR,AUC= %0.4f)' % auc_crp)
plt.plot(fpr_crp,tpr_crp,color='#FFCC66',label='CRIP(AUC= %0.4f)' % auc_crp)


"""CSCRSites"""


auc_gmnn = roc_auc_score(labels.values[6], preds.values[6])
fpr_gmnn, tpr_gmnn, thresholds = roc_curve(labels.values[6], preds.values[6])
plt.plot(fpr_gmnn,tpr_gmnn,color='#9F79EE',label='CSCRSites(AUC= %0.4f)' % auc_gmnn)


plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([-0.05, 1])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate',fontsize=27,fontproperties='Times New Roman')
plt.ylabel('True Positive Rate',fontsize=27,fontproperties='Times New Roman')
plt.yticks(fontproperties='Times New Roman', size=27)
plt.xticks(fontproperties='Times New Roman', size=27)

plt.title('FASTKD2_K562 using HepG2 model', size=30, fontproperties='Times New Roman')

plt.rcParams.update({'font.size':27})
plt.legend(loc='lower right',prop={"family": "Times New Roman", "size": 27})
plt.tight_layout()
plt.show()
```
